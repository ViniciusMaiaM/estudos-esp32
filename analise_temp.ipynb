```python
import pandas as pd
```

```python
df = pd.read_csv("feeds.csv")
```

```python
df.head()
```

```python
def corrige(string):

    new_str = ''.join([string[i] for i in range(len(string)) if i <= 18])

    return new_str 
    
```

```python
df["created_at"] = list(map(corrige, df["created_at"]))
```

```python
df.head()
```

```python
df["created_at"] = pd.to_datetime(df["created_at"])
df.dtypes
```

```python
df.head()
```

```python
#removendo colunas desncessarias
df = df.drop(columns=["latitude", "longitude", "elevation", "status"])
```
